### Analysis of Transactions Derived from the latest 100 fetched blocks


```python
import csv
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Read transactions from the CSV file into a pandas DataFrame
transactions_df = pd.read_csv('transactions.csv')
```

```python
transactions_df.head(10)
```

### Transactions Summaries Statistics

```python
transactions_df.describe()
```

Total Ether Transferred

```python
total_ether_transferred = transactions_df['value'].sum()
print(f'Total Ether transferred is {total_ether_transferred} ETH')
```

AVERAGE GAS PRICE

```python
average_gas_price = transactions_df['gas_price'].mean()
print(f'Average gas price for the transactions {average_gas_price} Gwe')
```

Total gas used

```python
total_gas_used = transactions_df['gas'].sum()
print(f"Total Gas Used: {total_gas_used} units")
```

### Visualizations

```python
transactions_df['block_timestamp_utc'] = pd.to_datetime(transactions_df['block_timestamp_utc'])

# Plot Gas Price Trend
plt.figure(figsize=(10, 6))
sns.lineplot(data=transactions_df, x='block_timestamp_utc', y='gas_price')
plt.title("Gas Price Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Gas Price (Gwei)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plot Ether Transferred Trend
plt.figure(figsize=(10, 6))
sns.lineplot(data=transactions_df, x='block_timestamp_utc', y='value')
plt.title("Ether Transferred Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Ether Transferred")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Plot Total Ether Transferred per Block
total_ether_per_block = transactions_df.groupby('block_number')['value'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=total_ether_per_block, x='block_number', y='value')
plt.title("Total Ether Transferred per Block")
plt.xlabel("Block Number")
plt.ylabel("Ether Transferred")
plt.tight_layout()
plt.show()
```

```python
# Plot Distribution of Gas Used (Histogram)
plt.figure(figsize=(10, 6))
sns.histplot(transactions_df['gas'], kde=True, color='purple', bins=30)
plt.title("Distribution of Gas Used in Transactions")
plt.xlabel("Gas Used")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# Highlight Large Transfers on the Ether Transferred Trend
plt.figure(figsize=(10, 6))
sns.lineplot(data=transactions_df, x='block_timestamp_utc', y='value', label="Ether Transferred", color='g')
```
